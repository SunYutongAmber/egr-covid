```python

import matplotlib.pyplot as plt
import numpy as np
import math
 
x = np.arange(0, 10000000, 0.1)

y1 = -23*np.exp( -x/23 ) + 23.5*x + 1000000000
y2 = -20*np.exp( -x/20 ) + 19.8*x + 1000000000
y3 = -17*np.exp( -x/17 ) + 9.5*x + 1000000000
 
plt.title("Total Cost Function")
plt.xlabel("Number of Vaccine")
plt.ylabel("Total Cost")
plt.plot(x,y1,label='Pfizer')
plt.plot(x,y2,label='AstraZeneca')
plt.plot(x,y3,label='Jessen')
plt.legend()
plt.grid()
plt.show()

 
```
