```python

import matplotlib.pyplot as plt
import numpy as np
import math
 
x = np.arange(0, 1000, 0.1)

y1 = np.exp( -x/23 ) + 23.5
y2 = np.exp( -x/20 ) + 19.8
y3 = np.exp( -x/17 ) + 18.7
 
plt.title("Marginal Cost Function")
plt.xlabel("Number of Vaccine")
plt.ylabel("Marginal Cost")
plt.plot(x,y1,label='Pfizer')
plt.plot(x,y2,label='AstraZeneca')
plt.plot(x,y3,label='Johnson')
plt.legend()
plt.grid()
plt.show()

 
```
